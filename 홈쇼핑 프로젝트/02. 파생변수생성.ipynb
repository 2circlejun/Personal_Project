# 파생변수 생성
-  5월 30일에 데이터 특성을 확인해 본 결과 한 방송에서 여러 제품이 판매되는데, 각 상품별로 목표주문금액이 설정되어 있던 것을 황시율 조장이 발견했습니다.
-  => 예시를 들기 위해 한 방송에서 조리도구들을 판매할 때 총 1억원 어치의 목표주문금액이 있다고 가정하겠습니다
-  Product Dataset에서는, 방송의 총 목표 금액을 상품 종류의 개수로 나눠서 각각 품목에 할당되어 있습니다.
-  예를 들어 주방용품 판매 방송에서 국자, 냄비, 후라이팬, 접시, 총 네 가지 종류가 소개되었다면, 각각의 품목에 2500만원의 목표 주문금액이 설정되어 있습니다.
-  그러므로 쇼호스트가 방송에 국자만 2억원 어치를 판매하여 목표달성금액을 충족해도, 나머지 세 품목에서 각각 2,500만원어치의 매출이 발생하지 않으면 해당 방송의 쇼호스트 목표 달성률은 0.25가 되는 것 입니다.
