```python
import pandas as pd
import matplotlib.pyplot as plt

# '프로그램명'과 관련된 열을 선택하여 데이터프레임 생성
program_df = df[['프로그램명', '상품순주문수량', '목표달성여부']]

# '프로그램명'을 기준으로 그룹화하여 집계 함수 적용
grouped = program_df.groupby('프로그램명').agg({'상품순주문수량': 'count', '목표달성여부': 'sum'})

# 그래프 생성
fig, ax = plt.subplots(figsize=(20, 30))
grouped.plot(kind='barh', ax=ax)

# 그래프 설정
ax.set_xlabel('평균 / 합계')
ax.set_ylabel('프로그램명')
ax.set_title('프로그램별 상품순주문수량 평균 및 목표달성여부 합계')

# 그래프 출력
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

top_n = 10  # Number of top programs to display

# Count of goal achievements
top_programs_achievements = df.groupby('프로그램명')['목표달성여부'].sum().nlargest(top_n)

# Count of broadcasts
top_programs_broadcasts = df.groupby('프로그램명')['방송Code'].count().nlargest(top_n)

# Combine the data into a single DataFrame
combined_data = pd.concat([top_programs_achievements, top_programs_broadcasts], axis=1)
combined_data.columns = ['목표달성 수', '방송 횟수']

# Plotting
fig, ax = plt.subplots()
combined_data.plot(kind='bar', ax=ax)

# Add the total count as text above each bar
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{int(height)}', (p.get_x() + p.get_width() / 2, height),
                ha='center', va='bottom')

plt.xlabel('프로그램명')
plt.ylabel('방송 횟수')
plt.title(f'Top {top_n} 프로그램별 목표달성 수 및 방송 횟수')

# Customize x-axis tick labels
xticks = np.arange(len(combined_data.index))
plt.xticks(xticks, combined_data.index, rotation=45, ha='right')

plt.legend()
plt.tight_layout()
plt.show()
```
